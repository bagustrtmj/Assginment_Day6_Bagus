##### Tugas/Soal:

Temukan kota-kota dari worldcity yang memenuhi semua syarat yang tertulis pada poin-poin berikut ini:
- populasi kota tersebut melebih populasi dari kota terbanyak penduduknya di negara Filipina. 
- (Maksudnya: misalnya kota X adalah kota di Filipina yang paling banyak penduduknya. Kita mau daftar kota-kota dari worldcity yang jumlah penduduknya lebih besar dari jumlah penduduk di kota X tersebut)
- ada kota di Amerika Serikat yang memiliki nama yang sama dengan kota-kota pilihan kita dari worldcity (seperti kota Cave, Bergen, dll)

Keluarkan hanya nama kota-kota yang memenuhi 2 syarat di atas saja, tidak usah kolom-kolom lain.

##### NOTE: Hilangkan duplicate rows.

```python
query_1 = """
SELECT DISTINCT city, country, population
FROM worldcity
WHERE population > (
  SELECT max(population) country 
  FROM worldcity 
  WHERE country = 'Philippines'
  )
ORDER BY population ASC
"""

sql_run(query_1)
```

```python
query_2 = """
SELECT wc.city
FROM worldcity as wc
JOIN uscity as uc
ON wc.city = uc.city
WHERE wc.population > 23088000
"""

sql_run(query_2)
```

### Question 2: US Cities

Temukan kota-kota (city) di Amerika Serikat yang memenuhi seluruh poin-poin di bawah ini:
- memiliki nama 'city' yang tidak sama dengan 'state_name'
- city tersebut memiliki populasi di atas rata-rata populasi kota-kota di 'county_name' Miami-Dade
- terdiri dari dua atau lebih kata (contoh: Los Angeles, New York)

Output yang diharapkan hanya kolom 'city' saja, tidak usah kolom yang lain

```python
query_3 = """
WITH cns as(
SELECT uc.city , uc.state_name , uc.county_name, uc.population 
FROM uscity as uc 
WHERE uc.city != uc.state_name),

am AS(
SELECT ROUND(AVG(population)) as pop_miami, city, population, county_name 
FROM uscity as uc  
WHERE county_name = 'Miami-Dade'
GROUP BY city  
),

pop AS (
SELECT uc.city, ROUND(AVG(uc.population)) as total_population
FROM uscity as uc, am
WHERE uc.county_name != 'Miami-Dade'
GROUP BY uc.city 
HAVING total_population > am.pop_miami
)

SELECT CASE
	WHEN instr(pop.city, ' ') > 0 then pop.city
	END AS Cities
FROM pop
WHERE Cities NOTNULL ;
"""

sql_run(query_3)
```
